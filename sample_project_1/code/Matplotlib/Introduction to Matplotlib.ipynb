# Matplotlib Example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) #(x,y)

# 3. Plot data
ax.plot(x,y);

# Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & Show
fig.savefig("../../Images/sample.png")
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
# Plot to each different axis
ax1.plot(x,np.sin(x));
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
fig.savefig("../../Images/sample2.png")
```

```python
# Subplots otion 2
fig, ax = plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method?)
- When plotting something quickly, okay to use the pyplot method
- When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            cmap="winter");
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
             cmap="winter",
            ax = ax );
# ax.set_xlim([45,100]);
```

```python
## OO methom from scrath
fig, ax = plt.subplots(figsize=(10,6))
# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")

# Cutomize the plot
ax.set(title="Heart Disease and Cholestrol levels",
      xlabel="Age",
      ylabel="Cholestrol")
# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")
# Add a Horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
ax.axhline(over_50["chol"].min(),
           linestyle='--');
ax.axhline(over_50["chol"].max(),
           linestyle='--');
```

```python
over_50[:5]
```

```python
# Subplot of chol, age,thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10))
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")
# Change the x-axis limit
ax0.set_xlim([50,80])

# Customize ax0

ax0.set(title="Heart Disease and Cholestrol levels",
      xlabel="Age",
      ylabel="Cholestrol")
# Add legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target") 
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# Add data to ax1

scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")
# Change the x-axis & y-axis limit
ax1.set_xlim([50,80])
ax1.set_ylim([67,200])
# Customize ax1

ax1.set(title="Heart Disease and thalach levels",
      xlabel="Age",
      ylabel="Max Heart Rate")

# Add legend to ax1

ax1.legend(*scatter.legend_elements(),title="Target") 

# Add a meanline

ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add the titel to figure
fig.suptitle("Heart Disease Analysis" ,fontsize=50,fontweight="bold",fontname="cursive");

# Exporting figure
fig.savefig("heart-disease-analysis-plot")
```
